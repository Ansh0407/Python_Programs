IMPORTING LIBRARIES

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importing Dataset

```python
df = pd.read_csv("DATASET/University.csv")
```

INFORMATION ON THE DATASET

```python
df.info()
```

```python
print(df.head(10))
```

LINE GRAPH

```python
df2 = df.head(10)
plt.plot(df2["world_rank"], df2["institution"])
plt.xlabel('world_rank')
plt.ylabel('institution')
plt.title('Rank', fontsize = 20)
plt.show()
```

Bar Graph

```python
plt.bar(df2["world_rank"], df2["institution"])
plt.xlabel('world_rank')
plt.ylabel('institution')
plt.title('Rank', fontsize = 20)
plt.show()
```

Scatter plot

```python
plt.scatter(df2["citations"], df2["year"])
plt.xlabel('citations')
plt.ylabel('year')
plt.title('Scatter', fontsize = 20)
plt.show()
```

Correlation between National Rank and World Rank 

```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
import seaborn as sns


plt.figure(figsize=(12, 8)) 
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

plt.title("Correlation Heatmap")
plt.show()
```
